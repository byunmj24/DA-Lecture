## 순열(Permutations) 과 조합(Combinations)

- itertools.permutations(데이타리스트, n)
- itertools.combinations(데이타리스트, n)

```python
# 내장 모듈 - 아마 있을 것...
from itertools import permutations
```

```python
# 카드 1, 2, 3 으로 구성할 수 있는 경우의 수
lists = [1, 2, 3]
permutations(lists, 2)
```

```python
list(permutations(lists, 2))
```

```python
for item in permutations(lists, 2):
    print(item)
```

```python
len(list(permutations(lists, 2)))
```

```python
from itertools import combinations
```

```python
# 카드 3장에서 2장을 선택한다.
cards = [1, 2, 3]
list(combinations(cards, 2))
```

## 퀴즈
한 반의 학생 수가 40명입니다.

(1) 반장 1명, 부반장 1명을 뽑는 경우의 수를 구하세요.

(2) 주번 3명을 뽑는 경우의 수를 구하세요.

### (1)

```python
students = list(np.arange(1, 41))
```

```python
per_stu = list(permutations(students, 2))
```

```python
for stu in per_stu:
    print(stu)
```

```python
len(per_stu)
```

### (2)

```python
com_stu = list(combinations(students, 3))
```

```python
for stu in com_stu:
    print(stu)
```

```python
len(com_stu)
```
