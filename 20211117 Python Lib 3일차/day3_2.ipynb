### 조건 필터링

- 비교나 논리연산자를 이용해서 데이터 추출
- Boolean Index : 배열[조건식]
```
: 배열[조건식]
: 논리연산자 ~(not), |(or), &(and)
```
- np.where(조건식)

### Boolean Index

````python
arr1
````

````python
# 30 이하의 숫자만 필터링
arr1 <= 30
````

````python
arr1[arr1 <= 30]
````

````python
# 50 ~ 70 사이의 숫자만 필터링
(arr1 >= 50) & (arr1 <= 70)
````

````python
arr1[(arr1 >= 50) & (arr1 <= 70)]
````

````python
# 20보다 작거나 70보다 큰 숫자 필터링
(arr1 <= 20) | (arr1 >= 70)
````

````python
arr1[(arr1 <= 20) | (arr1 >= 70)]
````

````python
# 20보다 작거나 70보다 큰 숫자 제외
~((arr1 <= 20) | (arr1 >= 70))
````

````python
arr1[~((arr1 <= 20) | (arr1 >= 70))]
````

### np.where()

````python
arr2 = np.random.randint(1, 100, 10)
arr2
````

````python
# 70보다 크다 => 인덱스가 반환
np.where(arr2 > 70)
````

````python
arr2 > 70
````

````python
# 배열 실제값 반환 = 배열[np.where(조건식)]
arr2[np.where(arr2 > 70)]
````

````python
# 90보다 크거나 10보다 작은 데이터 출력
np.where((arr2 > 90) | (arr2 < 10))
````

````python
arr2[np.where((arr2 > 90) | (arr2 < 10))]
````

## 상관계수( Correlation Coefficient)
- 두 변수 사이의 상관관계의 정도를 수치적으로 나타낸 계수
- 피어슨 상관계수
- -1 에서 1사이의 범위를 갖는다.
- 0 이면 상관관계 없음으로 예측
- np.corrcoef(x, y)

````python
x = [67., 87., 90., 98., 45., 45., 56., 98.]
y = [78., 76., 98., 89., 95., 59., 96., 88.] 
````

````python
np.corrcoef(x, y)
````

````python
np.corrcoef(x, y)[0, 1], np.corrcoef(x, y)[1, 0]
````
