# 파이썬의 xml 

- xmltodict 라이브러리 
- pip3 install xmltodict

````python
import requests
import xmltodict
import json
import pandas as pd
# from xml.etree.ElementTree import *
````

````python
with open('country_data.xml', 'r') as file:
    xml_string = file.read()
````

````python
# text => OrderedDict
temp = xmltodict.parse(xml_string)
# type(temp), temp
````

````python
# 2) 리스트 딕셔너리 구조화 
# JSON 형태로 변경하여 Dictionary로 변경
dict = json.loads(json.dumps(temp))
type(dict), len(dict)
````

````python
dict['data']['country']
````

````python
df_country = pd.DataFrame(dict['data']['country'], columns=['@name', 'rank', 'year', 'gdppc'])
df_country
````

````python
# 1) 페이지 요청 

url = 'http://openapi.data.go.kr/openapi/service/rest/Covid19/getCovid19GenAgeCaseInfJson'
# 서비스키 설정 - 디코딩한 서비스키 이용 
serviceKey = 'Yhq6n9vlpaKZsCOw0jec8zIdZ8p+Bpuku6WLjVgDhqXRW6dHnfXoauSEj19jpnjv59CUGspyTxVCgCeXOua7dg=='

params ={'serviceKey' : serviceKey, 
         'pageNo' : '1', 
         'numOfRows' : '10', 
         'startCreateDt' : '20200101', # 시작일
         'endCreateDt' : '20211118' } # 종료일

response = requests.get(url, params=params)
xml_string = response.text
print(type(xml_string))
# print(xml_string)
````

````python
# 2) 리스트 딕셔너리 구조화 
# text => OrderedDict
temp = xmltodict.parse(xml_string)
# JSON 형태로 변경하여 Dictionary로 변경
dict = json.loads(json.dumps(temp))
type(dict), len(dict)
````
