```python
def lsm(X, Y):
    mean_X = np.mean(X)
    mean_Y = np.mean(Y)
    
    diff_X = sum([(i-mean_X)**2 for i in range(len(X))])
    
    diff_value = 0
    for i in range(len(X)):
        diff_value += (X[i] - mean_X) * (Y[i] - mean_Y)
        
    global a 
    global b
    
    a = diff_value / diff_X
    b = mean_Y - (mean_X * a)
    
    return a, b
```

```python
lsm(alcohol, quality)
```

```python
def l_reg(X, Y, lr, epochs):
    global a, b
    
    lsm(X, Y)
    print('ìµœì´ˆ ê¸°ìš¸ê¸° a:', a, 'yì ˆíŽ¸ b:', b)
    for i in range(epochs):
        predict_weight = a * X + b
        err = Y - predict_weight

        a_diff = -(1/len(X)) * sum(X * err)
        b_diff = -(1/len(X)) * sum(err)

        a = a - lr * a_diff
        b = b - lr * b_diff

        if i % 500 == 0:
            print('epoch=%.f, ê¸°ìš¸ê¸°=%.04f, ì ˆíŽ¸=%.04f' %(i, a, b))
```
