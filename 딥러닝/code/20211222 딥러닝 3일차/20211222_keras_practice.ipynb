```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import time
```

```python
x = np.arange(-1, 1, 0.01)
np.random.shuffle(x)
y = np.sin(10*x) + 0.5 * np.cos(5*x)

plt.scatter(x, y)
plt.show()
```

```python
index = int(len(x)*0.8)

train_x, test_x = x[:index], x[index:]
train_y, test_y = y[:index], y[index:]
```

```python
import tensorflow as tf
from tensorflow import keras

from tensorflow.keras import optimizers
from tensorflow.keras.layers import Dense, Input
```

```python
model = keras.Sequential()
model.add(Input(1))
model.add(Dense(10, activation='tanh'))
model.add(Dense(10, activation='tanh'))
model.add(Dense(1))

model.compile(loss='mse', optimizer='Adam', metrics=['mae'])
```

```python
hist = model.fit(train_x, train_y, epochs=1000*10, batch_size=128, validation_split=0.1, verbose=0)
```

```python
plt.plot(hist.history['loss'])
plt.plot(hist.history['val_loss'])
plt.show()
```

```python
plt.plot(hist.history['mae'])
plt.plot(hist.history['val_mae'])
plt.show()
```

```python
y_ = model.predict(test_x)
plt.scatter(test_x, test_y)
plt.scatter(test_x, y_)
plt.show()
```

# 다양한 입출력

## 2개의 입력, 1개의 출력

x1 + x2 -> y의 함수를 학습

x1과 x2는 0~1의 범위를 갖는다.

```python
x = np.ones((100, 2))
x[:, 0] = np.random.rand(100)
x[:, 1] = np.random.rand(100)
y = np.sum(x, axis=1)
```

```python
index = int(x.shape[0]*0.8)
train_x, test_x = x[:index], x[index:]
train_y, test_y = y[:index], y[index:]
```

```python
model = keras.Sequential()
model.add(Input(2))
model.add(Dense(10, activation='tanh'))
model.add(Dense(10, activation='tanh'))
model.add(Dense(1))
```

```python
model.compile(optimizer='SGD', loss='mse', metrics=['mae'])
model.summary()
```

```python
start = time.time()
model.fit(train_x, train_y, epochs=1000*5, verbose=0, batch_size=20)
print('elapsed :', (time.time()-start))
```

```python
loss, mae = model.evaluate(test_x, test_y)
print("loss=", loss)
print("mae=", mae)
```

```python
y_ = model.predict(test_x)
```

```python
plt.plot(test_y, 'o')
plt.plot(y_, '.')
plt.show()
```
