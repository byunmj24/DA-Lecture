```python
import pandas as pd

train = pd.read_csv(os.path.join(PATH, 'train.csv'))
```

```python
train.columns
```

```python
train_columns = list(train.columns)[1:-1]
```

```python
train_columns
```

```python
data = train[train_columns]
```

```python
imputer = SimpleImputer(strategy='mean')
data = imputer.fit_transform(data)
```

```python
scaler = StandardScaler()
data = scaler.fit_transform(data)
```

```python
train[train_columns] = data
```

```python
train
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_data = data
Y_data = train['Cover_Type'].values
```

```python
X_data.shape
```

```python
X_train, X_valid, Y_train, Y_valid = train_test_split(X_data, Y_data, test_size=0.2, stratify=Y_data)
# stratify=Y_data 
# Y_data에 대해서 같은 분포로 샘플링 나눠줌
```

```python
pd.Series(Y_train).describe()
```

```python
pd.Series(Y_valid).describe()
```

```python
X_train.shape
```

```python
Y_train.shape
```

## Multi Class LogisticRegression

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
```

```python
lr.fit(X_train, Y_train)
```

```python
pred_valid = lr.predict(X_valid)
```

```python
pred_valid
```

```python
from sklearn.metrics import accuracy_score,f1_score
```

```python
accuracy_score(pred_valid, Y_valid)
```

```python
f1_score(pred_valid, Y_valid, average='macro')
```

```python
f1_score(pred_valid, Y_valid, average='micro')
```

## 데이터 전처리

### Null 값 처리

```python
train_data.info()
```

```python
train_data.isna().sum()
```

```python
train_data.dropna(inplace=True)
```

```python
train_data.info()
```

### Train, Target 데이터 분리

```python
train_data.columns
```

```python
train_columns = ['Elevation', 'Aspect', 'Slope',
       'Horizontal_Distance_To_Hydrology', 'Vertical_Distance_To_Hydrology',
       'Horizontal_Distance_To_Roadways', 'Hillshade_9am', 'Hillshade_Noon',
       'Hillshade_3pm', 'Horizontal_Distance_To_Fire_Points', 'Soil_Type',
       'Wilderness_Area']
```

```python
x_data = train_data[train_columns]
y_data = train_data['Cover_Type'].values
```

```python
x_data
```

### Categorical Type Encoding

```python
x_data['Soil_Type'] = x_data['Soil_Type'].astype('category').cat.codes
x_data['Wilderness_Area'] = x_data['Wilderness_Area'].astype('category').cat.codes
```

```python
x_data = pd.get_dummies(x_data, columns=['Soil_Type'])
x_data = pd.get_dummies(x_data, columns=['Wilderness_Area'])
```

```python
x_data
```

```python
scaled_columns = [
 'Elevation',
 'Aspect',
 'Slope',
 'Horizontal_Distance_To_Hydrology',
 'Vertical_Distance_To_Hydrology',
 'Horizontal_Distance_To_Roadways',
 'Hillshade_9am',
 'Hillshade_Noon',
 'Hillshade_3pm',
 'Horizontal_Distance_To_Fire_Points']
```

```python
scaler = StandardScaler()
x_data[scaled_columns] = scaler.fit_transform(x_data[scaled_columns])
```

```python
x_data = x_data.to_numpy()
```

```python
X_train, X_valid, Y_train, Y_valid = train_test_split(x_data, y_data, test_size=0.2, stratify=y_data)
```
